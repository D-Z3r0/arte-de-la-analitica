# Conclusin
A partir de la manipulacin de los datos, se puede observar que estos estn relacionados con la produccin y distribucin de aguacates en Estados Unidos, ya que diversas variables de los datos revelan esta informacin. Primeramente, la variable de la fecha y el ao pueden indicar la fecha en se distribuyen los aguacates. A su vez, el ao tiene una fuerte correlacin con el precio promedio, por lo que se puede interpretar que dependiendo ao, el precio del aguacate vara. En cuanto a las variables "Total Volume", "4046","4225" y "4770", con base en sus correlaciones se puede establecer que el volumen total es calculado a partir de la suma total de la cantidades de los diferentes tipos de aguacates. Asimismo, cada tipo de aguacate tiene su propio valor segn la fecha, por lo que indica la cantidad que se distribuy de ese tipo de aguacate. Por otro lado, para las variables "Total Bags", "Small Bags", "Large Bags" y " XLarge Bags" se puede suponer, a partir de las correlaciones, que las cantidades de estas se relacionan con la cantidad de cada tipo de aguacate. Esta suposicin se debe a que cada tipo de aguacate tiene sus propias caractersticas que requieren de ser empaquetadas en bolsas pequeas, largas o extralargas, por ejemplo, el aguacate de tipo 4770 se conoce por ser ms grande, por lo que la cantidad de bolsas va a variar segn el tamao en las que vayan a ser empaquetadas. Por ltimo, la variable de la regin indica el estado a donde fueron distribuidas cierta cantidad de aguacates.
