# Conclusión
A partir de la manipulación de los datos, se puede observar que estos están relacionados con la producción y distribución de aguacates en Estados Unidos, ya que diversas variables de los datos revelan esta información. Primeramente, la variable de la fecha y el año pueden indicar la fecha en se distribuyen los aguacates. A su vez, el año tiene una fuerte correlación con el precio promedio, por lo que se puede interpretar que dependiendo año, el precio del aguacate varía. En cuanto a las variables "Total Volume", "4046","4225" y "4770", con base en sus correlaciones se puede establecer que el volumen total es calculado a partir de la suma total de la cantidades de los diferentes tipos de aguacates. Asimismo, cada tipo de aguacate tiene su propio valor según la fecha, por lo que indica la cantidad que se distribuyó de ese tipo de aguacate. Por otro lado, para las variables "Total Bags", "Small Bags", "Large Bags" y " XLarge Bags" se puede suponer, a partir de las correlaciones, que las cantidades de estas se relacionan con la cantidad de cada tipo de aguacate. Esta suposición se debe a que cada tipo de aguacate tiene sus propias características que requieren de ser empaquetadas en bolsas pequeñas, largas o extralargas, por ejemplo, el aguacate de tipo 4770 se conoce por ser más grande, por lo que la cantidad de bolsas va a variar según el tamaño en las que vayan a ser empaquetadas. Por último, la variable de la región indica el estado a donde fueron distribuidas cierta cantidad de aguacates.
